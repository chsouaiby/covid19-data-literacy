```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cases_deaths=pd.read_csv('WHO-COVID-19-global-data.csv')
icubeds=pd.read_csv('current-covid-patients-icu.csv')
```

## $\color{green}{\text{IT}}$$\color{black}{\text{A}}$$\color{red}{\text{LY}}$ 

```python
italy_cd=cases_deaths[(cases_deaths['Country']=='Italy')]
italy_cd=italy_cd[['Country','Date_reported','New_cases','New_deaths']]

italy_icu=icubeds[(icubeds['Entity']=='Italy')]
italy_icu=italy_icu[['Entity','Day','Daily ICU occupancy']]
```

```python
italy_cd=italy_cd.reset_index()
italy_icu=italy_icu.reset_index()
```

```python
italy_cd[(italy_cd['Date_reported']=='2020-02-24')]
italy_cd=italy_cd[52:]
italy_cd
```

```python
italy_cd=italy_cd.reset_index()
italy_cd
```

```python
italy_icu=italy_icu[:801]
italy_icu
```

```python
italy=italy_cd.join(italy_icu['Daily ICU occupancy'],how='right')
italy
```

```python
italy1=italy[(italy['Date_reported']=='2021-01-01')]
italy1
```

```python
italy2=italy[(italy['Date_reported']=='2022-01-01')]
italy2
```

```python
italy2020_keydates=['2020-03-04','2020-05-04','2020-10-13','2020-11-03']
italy2021_keydates=['2021-04-22','2021-06-28','2021-08-28']
italy2022_keydates=['2022-04-30']
```

## $\color{green}{\text{2020}}$ 

```python
italy2020=italy[:312]
italy2020
```

```python
ita_dates_2020=italy2020['Date_reported']
italy_dates_2020=[ita_dates_2020[i] for i in range(312) if i%7==0 or ita_dates_2020[i] in italy2020_keydates]
```

```python
italy_cases_2020=italy2020['New_cases']
italy_cases_2020=[italy_cases_2020[i] for i in range(312) if ita_dates_2020[i] in italy_dates_2020]
```

```python
italy_deaths_2020=italy2020['New_deaths']
italy_deaths_2020=[italy_deaths_2020[i] for i in range(312) if ita_dates_2020[i] in italy_dates_2020]
```

```python
italy_icu_2020=italy2020['Daily ICU occupancy']
italy_icu_2020=[italy_icu_2020[i] for i in range(312) if ita_dates_2020[i] in italy_dates_2020]
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid ICU beds in Italy in 2020')
plt.plot(italy_dates_2020,italy_icu_2020,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.annotate("March 4:\n1st Lockdown",('2020-03-04',295));
plt.annotate("May 4:\nGradual\nlifting of\nrestrictions",('2020-05-04',1479));
plt.annotate("October 13:\nReimposing\nrestrictions",('2020-10-13',514));
plt.annotate("November 3:\n3 level\nalert system",('2020-11-03',2225));
plt.legend();
plt.savefig("italyICU2020.png",dpi=300);
plt.show();
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid in Italy in 2020')
plt.plot(italy_dates_2020,italy_cases_2020,label="New Cases",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(italy_dates_2020,italy_deaths_2020,label="New Deaths",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(italy_dates_2020,italy_icu_2020,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.annotate("March 4:\n1st Lockdown",('2020-03-04',3000));
plt.annotate("May 4:\nGradual\nlifting of\nrestrictions",('2020-05-04',1750));
plt.annotate("October 13:\nReimposing\nrestrictions",('2020-10-13',1500));
plt.annotate("November 3:\n3 level\nalert system",('2020-11-03',4000));
plt.legend();
plt.savefig("italy2020.png",dpi=300);
plt.show();
```
