```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cases_deaths=pd.read_csv('WHO-COVID-19-global-data.csv')
```

## $\color{blue}{\text{NE}}$$\color{black}{\text{W}}$ $\color{red}{\text{ZEA}}$$\color{black}{\text{LA}}$$\color{blue}{\text{ND}}$

```python
newzealand=cases_deaths[(cases_deaths['Country']=="New Zealand")]
newzealand=newzealand[['Date_reported','New_cases','New_deaths']]
```

```python
newzealand=newzealand.reset_index()
newzealand
```

```python
nz1=newzealand[(newzealand['Date_reported']=='2021-01-01')]
nz1
```

```python
nz2=newzealand[(newzealand['Date_reported']=='2022-01-01')]
nz2
```

```python
nz2020_keydates=['2020-02-28','2020-03-18','2020-03-24','2020-04-27','2020-05-18','2020-08-11','2020-10-07']
nz2021_keydates=['2021-02-14','2021-02-22','2021-02-28','2021-03-06','2021-08-17','2021-09-06','2021-12-03']
```

## $\color{blue}{\text{2020}}$ 

```python
nz2020=newzealand[:364]
```

```python
newzealand_dates_2020=nz2020['Date_reported']
nz_dates_2020=[newzealand_dates_2020[i] for i in range(364) if i%7==0 or newzealand_dates_2020[i] in nz2020_keydates]

nz_cases_2020=nz2020['New_cases']
nz_cases_2020=[nz_cases_2020[i] for i in range(364) if newzealand_dates_2020[i] in nz_dates_2020]

nz_deaths_2020=nz2020['New_deaths']
nz_deaths_2020=[nz_deaths_2020[i] for i in range(364) if newzealand_dates_2020[i] in nz_dates_2020]
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid in New Zealand in 2020')
plt.plot(nz_dates_2020,nz_cases_2020,label="New Cases",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(nz_dates_2020,nz_deaths_2020,label="New Deaths",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.annotate("February 28:\nFirst Case",('2020-02-28',3))
plt.annotate("March 18:\nInternational\ntravel ban",('2020-03-18',60))
plt.annotate("March 24:\nL4 National\nLockdown",('2020-03-24',70))
plt.annotate("April 27:\nL3 Alert",('2020-04-27',5))
plt.annotate("May 18:\nL1 Alert\nLockdown lift",('2020-05-18',2))
plt.annotate("August 11:\nL2 National\nLockdown\n(L3 Auckland)",('2020-08-11',15))
plt.annotate("October 7:\nAuckland\nback to L1",('2020-10-07',8))
plt.legend();
plt.savefig("newzealand2020.png",dpi=300);
plt.show();
```

## $\color{red}{\text{2021}}$ 

```python
nz2021=newzealand[364:729]
```

```python
newzealand_dates_2021=nz2021['Date_reported']
nz_dates_2021=[newzealand_dates_2021[i] for i in range(364,729) if i%7==0 or newzealand_dates_2021[i] in nz2021_keydates]

nz_cases_2021=nz2021['New_cases']
nz_cases_2021=[nz_cases_2021[i] for i in range(364,729) if newzealand_dates_2021[i] in nz_dates_2021]

nz_deaths_2021=nz2021['New_deaths']
nz_deaths_2021=[nz_deaths_2021[i] for i in range(364,729) if newzealand_dates_2021[i] in nz_dates_2021]
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid in New Zealand in 2021')
plt.plot(nz_dates_2021,nz_cases_2021,label="New Cases",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(nz_dates_2021,nz_deaths_2021,label="New Deaths",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.annotate("February 14:\nL2 National\nLockdown\n(L3 Auckland)",('2021-02-14',10))
plt.annotate("February 22:\nLockdown\nlift",('2021-02-22',50))
plt.annotate("February 28:\nL3 Lockdown\nAuckland",('2021-02-28',35))
plt.annotate("March 6:\nLockdown\nlift",('2021-03-06',75))
plt.annotate("August 17:\nL4 National Lockdown",('2021-08-17',75))
plt.annotate("September 6:\nLockdown lift\n(not Auckland)",('2021-09-06',50))
plt.annotate("December 3:\nProtection\nFramework\nreplaces alert\nlevel system",('2021-12-03',100))
plt.legend();
plt.savefig("newzealand2021.png",dpi=300);
plt.show();
```
