```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cases_deaths=pd.read_csv('WHO-COVID-19-global-data.csv')
```

```python
cases=cases_deaths[['Country','Date_reported','New_cases']]
```

```python
deaths=cases_deaths[['Country','Date_reported','New_deaths']]
```

```python
fra_cases=cases[(cases['Country']=='France')]
dates=np.array(fra_cases['Date_reported'])
fra_cases=np.array(fra_cases['New_cases'])
```

```python
ita_cases=cases[(cases['Country']=='Italy')]
ita_cases=np.array(ita_cases['New_cases'])
```

```python
nz_cases=cases[(cases['Country']=='New Zealand')]
nz_cases=np.array(nz_cases['New_cases'])
```

```python
sk_cases=cases[(cases['Country']=='Republic of Korea')]
sk_cases=np.array(sk_cases['New_cases'])
```

```python
vn_cases=cases[(cases['Country']=='Viet Nam')]
vn_cases=np.array(vn_cases['New_cases'])
```

```python
indi_cases=cases[(cases['Country']=='India')]
indi_cases=np.array(indi_cases['New_cases'])
```

```python
indo_cases=cases[(cases['Country']=='Indonesia')]
indo_cases=np.array(indo_cases['New_cases'])
```

```python
dates=[dates[i] for i in range(729) if i%14==0]
fra_cases=[fra_cases[i] for i in range(729) if i%14==0]
ita_cases=[ita_cases[i] for i in range(729) if i%14==0]
nz_cases=[nz_cases[i] for i in range(729) if i%14==0]
vn_cases=[vn_cases[i] for i in range(729) if i%14==0]
sk_cases=[sk_cases[i] for i in range(729) if i%14==0]
indi_cases=[indi_cases[i] for i in range(729) if i%14==0]
indo_cases=[indo_cases[i] for i in range(729) if i%14==0]
```

```python
plt.figure(figsize=[12.5,7]);
plt.title("Evolution of Covid-19 new cases in 7 countries");
plt.plot(dates,fra_cases,label='France',color='turquoise',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.plot(dates,ita_cases,label='Italy',color='green',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.plot(dates,nz_cases,label='New Zealand',color='fuchsia',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.plot(dates,vn_cases,label='Vietnam',color='yellow',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.plot(dates,sk_cases,label='South Korea',color='purple',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.plot(dates,indo_cases,label='Indonesia',color='red',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.plot(dates,indi_cases,label='India',color='orange',linewidth=3);
plt.xticks(rotation=45,ha='right');
plt.legend();
plt.savefig("WorldCases.png",dpi=600);
plt.show();
```
