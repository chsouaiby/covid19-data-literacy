```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cases_deaths=pd.read_csv('WHO-COVID-19-global-data.csv') #cases&deaths
```

```python
icubeds=pd.read_csv('current-covid-patients-icu.csv') #icubeds
```

## $\color{blue}{\text{FR}}$$\color{black}{\text{AN}}$$\color{red}{\text{CE}}$ 

This study covers dates from march 18 2020 to may 4 2022

```python
france_cd=cases_deaths[(cases_deaths['Country']=='France')]
france_cd=france_cd[['Country','Date_reported','New_cases','New_deaths']]

france_icu=icubeds[(icubeds['Entity']=='France')]
france_icu=france_icu[['Entity','Day','Daily ICU occupancy']]
```

```python
france_cd=france_cd.reset_index()
france_icu=france_icu.reset_index()
```

```python
france_cd=france_cd[75:]
```

```python
france_cd=france_cd.reset_index()
france_cd
```

```python
france_icu=france_icu[:778]
france_icu
```

```python
france=france_cd.join(france_icu['Daily ICU occupancy'],how='right')
france
```

```python
france1=france[(france['Date_reported']=='2021-01-01')]
france1
```

```python
france2=france[(france['Date_reported']=='2022-01-01')]
france2
```

```python
france2020_keydates=['2020-05-11','2020-10-30','2020-12-01']
france2021_keydates=['2021-01-01','2021-01-14','2021-01-29','2021-03-04','2021-03-18','2021-04-02','2021-04-23','2021-06-09','2021-07-13']
france2022_keydates=['2022-03-14','2022-05-04']
```

## $\color{blue}{\text{2020}}$ 

```python
france2020=france[:289]
france2020
```

```python
fra_dates_2020=france2020['Date_reported']
france_dates_2020=[fra_dates_2020[i] for i in range(289) if i%7==0 or fra_dates_2020[i] in france2020_keydates]
```

```python
france_cases_2020=france2020['New_cases']
france_cases_2020=[france_cases_2020[i] for i in range(289) if fra_dates_2020[i] in france_dates_2020]
```

```python
france_deaths_2020=france2020['New_deaths']
france_deaths_2020=[france_deaths_2020[i] for i in range(289) if fra_dates_2020[i] in france_dates_2020]
```

```python
france_icu_2020=france2020['Daily ICU occupancy']
france_icu_2020=[france_icu_2020[i] for i in range(289) if fra_dates_2020[i] in france_dates_2020]
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid ICU beds in France in 2020')
plt.plot(france_dates_2020,france_icu_2020,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.annotate("March 17:\n1st Lockdown", ('2020-03-18', 0));
plt.annotate("May 11:\nLockdown lift", ('2020-05-11', 2666));
plt.annotate("October 30:\n2nd Lockdown", ('2020-10-30', 3368));
plt.annotate("December 1:\n8pm curfew",('2020-12-01',3594));
plt.legend();
plt.savefig("franceICU2020.png",dpi=300);
plt.show();
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid in France in 2020')
plt.plot(france_dates_2020,france_cases_2020,label="New Cases",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(france_dates_2020,france_deaths_2020,label="New Deaths",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(france_dates_2020,france_icu_2020,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.annotate("March 17:\n1st Lockdown", ('2020-03-18', 9000));
plt.annotate("May 11:\nLockdown lift", ('2020-05-11', 4500));
plt.annotate("October 30:\n2nd Lockdown", ('2020-10-30', 6000));
plt.annotate("December 1:\n8pm curfew",('2020-12-01',4000));
plt.legend();
plt.savefig("france2020.png",dpi=300);
plt.show();
```

## ANALYSIS
For 2020, the lockodwn policies seem efficient enough, we can notice the lockdown is lifted as the number of ICU beds is decreasing, and is imposed again when both the number of ICU beds and the number of cases are increasing really fast, and even the number of deaths at that time seems to be increasing faster than usual(the small rise in the orange graph(**new** deaths)).

```python
plt.figure(figsize=[20,10])
plt.title('evolution of covid in France in 2021')
plt.plot(france_dates_2021,france_cases_2021,label="New Cases",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(france_dates_2021,france_deaths_2021,label="New Deaths",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(france_dates_2021,france_icu_2021,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.annotate("January 1:\nRegional\n6pm\ncurfew",('2021-01-01',25000));
plt.annotate("January 14:\nNational\n6pm\ncurfew",('2021-01-14',30000));
plt.annotate("January 29:\nBorders closed\nfor non EU\ncountries",('2021-01-29',25000));
plt.annotate("March 4:\nRegional\nweekend\nlockdown",('2021-03-03',35000));
plt.annotate("March 18:\nRegional\nfull\nlockdown",('2021-03-18',37500));
plt.annotate("April 2:\n3rd national\nlockdown",('2021-04-07',37500));
plt.annotate("April 23:\nlockdown\nlift",('2021-04-23',20000));
plt.annotate("June 9:\nReopening\ninternational\ntourism",('2021-06-09',5000));
plt.annotate("July 13:\ntightening\nHealth Pass\nrestrictions",('2021-07-13',25000));
plt.legend();
plt.savefig("france2021.png",dpi=300);
plt.show();
```

## ANALYSIS
As for 2021, We can clearly tell  the main strategy could be summarized in two words : "Avoid Lockdown". To what extent was this strategy efficient?
In the first months, the government tried in many ways (like regional curfews or weekend lockdowns) to contain the virus, but as the graph shows, this did not work so well. The number of ICU beds skyrocketed which means the pressure on the hospitals was out of control(as confirms my previous research).The french government was also criticized for starting too slow with the vaccination campaign.
It wasn't until after the nationwide lockdown that the situation started getting better as the number of ICU beds and the daily number of new cases started decreasing.
After that, no other lockdown was placed, the international tourism was even reopened, so the most logical explanation for the situation staying somehow under control would be the vaccination. The government insisted on the urge and the importance of the vaccination, and the health pass was used to encourage it and to some extent force it. Regardless being with or against the vaccine, it appears to be, according to the data, an efficient solution.
