```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cases_deaths=pd.read_csv('WHO-COVID-19-global-data.csv')
```

```python
icubeds=pd.read_csv('current-covid-patients-icu.csv')
```

## $\color{black}{\text{SOU}}$$\color{blue}{\text{TH}}$ $\color{red}{\text{KO}}$$\color{black}{\text{REA}}$

```python
southkorea_cd=cases_deaths[(cases_deaths['Country']=="Republic of Korea")]
southkorea_cd=southkorea_cd[['Date_reported','New_cases','New_deaths']]

southkorea_icu=icubeds[(icubeds['Entity']=='South Korea')]
southkorea_icu=southkorea_icu[['Entity','Day','Daily ICU occupancy']]
```

```python
southkorea_cd=southkorea_cd.reset_index()
southkorea_icu=southkorea_icu.reset_index()
```

```python
southkorea_cd[(southkorea_cd['Date_reported']=='2020-03-28')]
```

```python
southkorea_icu[(southkorea_icu['Day']=='5/4/2022')]
```

```python
southkorea_cd=southkorea_cd[85:]
```

```python
southkorea_cd=southkorea_cd.reset_index()
```

```python
southkorea_icu=southkorea_icu[:768]
```

```python
southkorea=southkorea_cd.join(southkorea_icu[['Entity','Daily ICU occupancy']],how='right')
southkorea
```

```python
sk1=southkorea[(southkorea['Date_reported']=='2021-01-01')]
sk1
```

```python
sk2=southkorea[(southkorea['Date_reported']=='2022-01-01')]
sk2
```

## $\color{blue}{\text{2020}}$ 

```python
sk2020=southkorea[:279]
sk2020
```

```python
sk_dates_2020=sk2020['Date_reported']
sk_dates_2020=[sk_dates_2020[i] for i in range(279) if i%7==0]
```

```python
sk_cases_2020=sk2020['New_cases']
sk_cases_2020=[sk_cases_2020[i] for i in range(279) if i%7==0]
```

```python
sk_deaths_2020=sk2020['New_deaths']
sk_deaths_2020=[sk_deaths_2020[i] for i in range(279) if i%7==0]
```

```python
sk_icu_2020=sk2020['Daily ICU occupancy']
sk_icu_2020=[sk_icu_2020[i] for i in range(279) if i%7==0]
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid ICU beds in South Korea in 2020')
plt.plot(sk_dates_2020,sk_icu_2020,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.legend();
plt.savefig("southkoreaICU2020.png",dpi=300);
plt.show();
```

```python
plt.figure(figsize=[16,8])
plt.title('evolution of covid in South Korea in 2020')
plt.plot(sk_dates_2020,sk_cases_2020,label="New Cases",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(sk_dates_2020,sk_deaths_2020,label="New Deaths",linestyle='dotted');
plt.xticks(rotation=45,ha='right');
plt.plot(sk_dates_2020,sk_icu_2020,label="Current ICU beds");
plt.xticks(rotation=45,ha='right');
plt.legend();
plt.savefig("southkorea2020.png",dpi=300);
plt.show();
```
